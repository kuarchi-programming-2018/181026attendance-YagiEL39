# ライブラリの`import`を用いる
## matplotlibを使ってグラフをプロット

まずはデータを読み込み，それをX, Yのデータにして返す関数を作りましょう。
csv とはcomma separated valuesです。テキストで開いてみればわかります。エクセルでも開けます。

```python
import csv
def read_data(file_name):
    reader = csv.reader(open(file_name,'r'))
    X=[]
    Y=[]
    for row in reader:
        X.append(float(row[0]))
        Y.append(float(row[1]))
    return X, Y  # データを返す
```

matplotlibを用いてデータを散布図にプロットする関数を作成します。

```python
import matplotlib.pyplot as plt
% matplotlib inline#  % matplotlib inlineはjupyter notebookでmatplotlibを行内に表示するためのおまじない
import numpy as np
```

```python
def draw_graph(X,Y,title,xlabel,ylabel):
    X=np.arange(-3,3,0.1)
    Y=np.sin(X)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.plot(X,Y)
    plt.show()  # グラフの描画
```

では，データを読み込んで描画して見ましょう。

```python
a='data3.csv'  # ファイル名
b,c=read_data(a)
draw_graph(b,c,"正弦曲線","X-LABEL","Y-LABEL") 
```
